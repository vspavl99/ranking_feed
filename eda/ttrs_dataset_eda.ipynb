## Correlation matrix

```python
corr_matrix = ttrs_dataset.corr()

sns.set(style='white')
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, cmap='seismic', vmin=-1, vmax=1, ax=ax)

# set the axis labels and title
ax.set_title('10x10 Correlation Matrix')

# display the plot
plt.show()
```
